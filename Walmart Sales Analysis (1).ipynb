```python
import pandas as pd
```

```python
walmart=pd.read_csv(r'C:\Users\DELL\Desktop\WalmartSalesData.csv')
```

```python
walmart.head()
```

```python
walmart.info()
```

```python
walmart['Date']=pd.to_datetime(walmart['Date'])
walmart.info()
```

```python
import dateutil

def get_part_of_day(hour):
     return (
        "morning" if 0 <= hour <= 12
        else
        "afternoon" if 12 <= hour <= 16
        else
        "evening" 
    )
walmart['part_of_day'] = walmart.Time.apply(lambda x: get_part_of_day(dateutil.parser.parse(x).hour)) 
```

```python
walmart.head()
```

```python
walmart['day_of_week'] = walmart['Date'].dt.day_name()
```

```python
walmart.head()
```

```python
walmart['month_name'] = walmart['Date'].dt.month_name()
```

```python
walmart.head()
```

```python
walmart.isna().sum()
```

```python
walmart.describe()
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
selling_product_line=(
    walmart
    .groupby('Product line')
    .agg({'Quantity':'sum',
           'Total':'sum'  })
    .sort_values('Quantity', ascending= True)
)
selling_product_line.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
(selling_product_line['Quantity']
.plot
 .barh(
    title='Quantity by Product Line',
     ylabel='Product Line',
     xlabel='Quantity',
     legend=None,
 )


)
plt.xticks(rotation=60)
sns.despine() 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
(selling_product_line['Total']
.plot
 .bar(
    title='Total Revenue by Product Line',
     ylabel='Total Revenue',
     xlabel='Product Line',
     legend=None,
    
 )
)
plt.xticks(rotation=60)
sns.despine() 
```

```python
import numpy as np
country_plot=walmart.pivot_table(index='month_name', values='Total', aggfunc=np.sum)
country_plot.sort_values('Total',ascending = True, inplace=True)
country_plot.plot(kind='line',figsize=(13,7), legend=None, )
plt.xlabel('Month')
plt.ylabel('Total')
plt.title('Total Revenue by Month')
plt.xticks(rotation=0)
plt.ylim(90000,120000)
plt.show();
```

```python
revenue_by_day=walmart.pivot_table(index='day_of_week', values='Total', aggfunc=np.sum)
revenue_by_day.sort_values('Total',inplace=True,ascending=False)
revenue_by_day.plot(kind='bar',figsize=(13,7), legend=False)
plt.xlabel('Day of Week')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Day of Week')
plt.xticks(rotation=0)
plt.show()
```

```python
branch_total_revenue=(
    walmart
    .groupby('Branch')
    .agg({'Total':'sum' })
    .sort_values('Total', ascending= False)
)
branch_total_revenue.head()
```

```python
(branch_total_revenue
.plot
 .bar(
    title='Total Revenue for each Branch',
     ylabel='Total Revenue',
     xlabel='Branch',
     legend=None,
    
 )
)
plt.xticks(rotation=0)
sns.despine() 
```

```python
saturday_sales=(
    walmart
    .query("day_of_week=='Saturday'")
)
```

```python
saturday_sales=(
    walmart
    .groupby('part_of_day')
    .agg({'Invoice ID':'count' })
    .sort_values('Invoice ID', ascending= False)
)
saturday_sales.head()
```

```python
(saturday_sales
.plot
 .bar(
    title='Saturday Sales',
     ylabel='No. of Sales',
     xlabel='Part of Day',
     legend=None,
    
 )
)
plt.xticks(rotation=0)
sns.despine() 
```

```python
customer_type_revenue=(
    walmart
    .groupby('Customer type')
    .agg({'Total':'sum'})
    .sort_values('Total', ascending=False)
)
customer_type_revenue.head()
```

```python
(customer_type_revenue
.plot
 .pie(
    title='Total Revenue by Customer Type',
     legend=None,
     subplots=True,
     
 )

)
plt.ylabel('')
plt.xticks(rotation=0)
sns.despine() 
```
